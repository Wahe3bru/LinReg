# Data Preperation

#### documenting the steps taken to transform multiple datasets into one to be used for modelling

1- transform data

2- prep for modelling



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
# data given
data = pd.read_csv('data_source/data.csv', parse_dates=True,infer_datetime_format= True)\
         .drop('Unnamed: 0', axis=1)
```

```python
data.head()
```

```python
debt = pd.read_csv('data_source/SA_Debt.csv', parse_dates=True,infer_datetime_format= True)
```

```python
debt.head()
```

```python
# Merge with debt and drop unnecessary columns
df_2 = data.merge(debt, how='left',
                 left_on=data['Date'].apply(lambda x: x[:4]),
                 right_on=debt['Date'].apply(lambda x: str(x)),
                 suffixes=('', '_y')).drop('Date_y', axis=1)
```

```python
#data + debt
df_2.head(1)
```

```python
cpi = pd.read_csv('data_source/SA_CPI.csv', parse_dates=True, infer_datetime_format=True)
```

```python
cpi.head()
```

```python
# Merge with cpi and drop unnecessary columns
df_3 = df_2.merge(cpi, how='left',
                 left_on=df_2['Date'].apply(lambda x: x[:4]),
                 right_on=cpi['year'].apply(lambda x: str(x))).drop('year', axis=1)
```

```python
# data + debt + cpi
df_3.head(1)
```

```python
gdp = pd.read_csv('data_source/SA_GDP.csv', parse_dates=True, infer_datetime_format=True)
```

```python
gdp.head()
```

```python
# Merge with gdp and drop unnecessary columns
df_4 = df_3.merge(gdp, how='left',
                 left_on=df_3['Date'].apply(lambda x: x[:4]),
                 right_on=gdp['Date'].apply(lambda x: str(x)),
                 suffixes=('', '_y')).drop('Date_y', axis=1)
```

```python
# data + debt + cpi + gdp
df_4.head(1)
```

```python
unemploy = pd.read_csv('data_source/SA_unemployment.csv', parse_dates=True, infer_datetime_format=True)
```

```python
unemploy.drop(['Unemployment','Unemployment_Change(%)'],axis=1,inplace=True)
```

```python
unemploy.head()
```

```python
# Merge with gdp and drop unnecessary columns
df_5 = df_4.merge(unemploy, how='left',
                 left_on=df_4['Date'].apply(lambda x: x[:4]),
                 right_on=unemploy['Date'].apply(lambda x: str(x)),
                 suffixes=('', '_y')).drop('Date_y', axis=1)
```

```python
# data + debt + cpi + gdp + unemployment
df_5.head(1)
```

```python
oil_df = pd.read_csv('data_source/BrentCrude.csv', delimiter=',',parse_dates=True, usecols=[2,3], names=['Date','Brent_Oil'], infer_datetime_format=True)
```

```python
oil_df['Date'] = oil_df['Date'].str.replace('-','/')
```

```python
# data + debt + cpi + gdp + unemployment + BrentCrude
oil_df.head()
```

```python
# Merge with BrentCrude
df_6 = df_5.merge(oil_df, how='left',
                 left_on=df_5['Date'],
                 right_on=oil_df['Date'],
                 suffixes=('', '_y')).drop('Date_y', axis=1)
```

```python
df_6.head()
```

```python
gold_df = pd.read_csv('data_source/gold.csv', delimiter=',',parse_dates=True, usecols=[2,3], names=['Date','Gold'])
```

```python
gold_df['Date'] = gold_df['Date'].str.replace('-','/')
```

```python
gold_df.head()
```

```python
# Merge with Gold and drop unnecessary columns
df_7 = df_6.merge(gold_df, how='left',
                 left_on=df_6['Date'],
                 right_on=gold_df['Date'],
                 suffixes=('', '_y')).drop('Date_y', axis=1)
```

```python
# data + debt + cpi + gdp + unemployment + BrentCrude + Gold
df_7.head()
```
